```python
import newspaper
```

```python
base_url = "https://www.cnn.com/"

articles = newspaper.build(base_url, language = 'en', memoize_articles = False)
```

```python
for article in articles.articles:
  print(article.url)
```

```python
import pandas as pd
```

```python
titles = []

for article in articles.articles:
  try:
    article.download()
    article.parse()
    
    article_features = {'title' : article.title,
                 'body' : article.text,
                 'url' : article.url}
    titles.append(article_features)
  except:
    print("Article Download Failure")


articles_df = pd.DataFrame.from_dict(titles)
articles_df.to_csv(r'CNN_Articles_May2_22.csv')
```
